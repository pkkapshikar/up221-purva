## Title: Profile of Census Tracts Along AC Transit 51A
#### Description: AC Transit 51A runs through several tracts in Oakland and Alameda, California. This file analyzes Census variables relating to demographics, housing, and transportation characteristics of these tracts.

```python
import pandas as pd
import geopandas as gpd
```

### Reading in data

```python
data = pd.read_csv("tracts_data_51a.csv")
```

```python
# data.head()
```

```python
data["FIPS"].dtype
```

### Join to CSV with geometry column

```python
geom_data = pd.read_csv("tracts_51A.csv")
```

```python
geom_data = geom_data.rename(columns={"GEOID": "FIPS"})
geom_data = geom_data.loc[:, ["FIPS", "geometry"]]
# geom_data.head()
```

```python
geom_data["FIPS"].dtype
```

Since both "FIPS" codes in the two dataframes are the same data type, we can merge the geometry column from the `geom_data` table to the initial `data`.

```python
joined_data = data.merge(right=geom_data, on="FIPS")
joined_data.head()
```

## Demographics variables

We first make the `joined_data` dataframe into a geodataframe.

```python
joined_data['geometry'] = gpd.GeoSeries.from_wkt(joined_data['geometry'])
geo_df = gpd.GeoDataFrame(joined_data, geometry='geometry')
```

#### Population Density

```python
geo_df.plot(figsize=(9,9),
            column="Population Density (Per Sq. Mile)",
            legend=True, 
            scheme='NaturalBreaks',
            legend_kwds={'loc': 'upper left'})
```

As the total populations by race/ethnicity were saved as counts and not percentages, I first created some percentages columns to plot.

```python
geo_df["Percent Black"] = geo_df["Total Population: Black or African American Alone"] / geo_df["Total Population"]
geo_df["Percent Asian"] = geo_df["Total Population: Asian Alone"] / geo_df["Total Population"]
geo_df["Percent White"] = geo_df["Total Population: White Alone"] / geo_df["Total Population"]
```

#### Percent Black population

```python
geo_df.plot(figsize=(9,9),
            column="Percent Black",
            legend=True, 
            scheme='NaturalBreaks',
            legend_kwds={'loc': 'upper left'})
```

#### Percent Asian population

```python
geo_df.plot(figsize=(9,9),
            column="Percent Asian",
            legend=True, 
            scheme='NaturalBreaks',
            legend_kwds={'loc': 'upper left'})
```

#### Percent White population

```python
geo_df.plot(figsize=(9,9),
            column="Percent White",
            legend=True, 
            scheme='NaturalBreaks',
            legend_kwds={'loc': 'upper left'})
```

As the variables for age are for certain ranges, I created a new column for ages 65 and older to plot.

```python
geo_df["Percent Total Population: 65 or More Years"] = (geo_df["Total Population: 65 to 74 Years"] + geo_df["Total Population: 75 to 84 Years"] + geo_df["Total Population: 85 Years and Over"]) / geo_df["Total Population"]
```

#### Percent population 65 years or older

```python
geo_df.plot(figsize=(9,9),
            column="Percent Total Population: 65 or More Years",
            legend=True, 
            scheme='NaturalBreaks',
            legend_kwds={'loc': 'upper left'})
```

## Housing Characteristics

#### Median Gross Rent

```python
geo_df.plot(figsize=(9,9),
            column="Median Gross Rent",
            legend=True, 
            scheme='NaturalBreaks',
            legend_kwds={'loc': 'upper left'})
```

#### Number of Workers 16 Years and Over in Renter-Occupied Housing Units using Public Transportation

```python
geo_df.plot(figsize=(9,9),
            column="Workers 16 Years and Over in Households: Householder Lived in Renter-Occupied Housing Units: Public Transportation (Excluding Taxicab)",
            legend=True, 
            scheme='NaturalBreaks',
            legend_kwds={'loc': 'upper left'})
```

I combined the variables for Renter-Occupied Housing Units where rent was 30-49% or 50% or more percent of income as these are rent-burdened households.

#### Percent Rent-Burdened Households out of all Renter-Occupied Housing Units

```python
geo_df["Percent Rent-Burdened"] = (geo_df["Renter-Occupied Housing Units: 30 to 49 Percent"] + geo_df["Renter-Occupied Housing Units: 50 percent or More"]) / geo_df["Renter-Occupied Housing Units"]
```

```python
geo_df.plot(figsize=(9,9),
            column="Percent Rent-Burdened",
            legend=True, 
            scheme='NaturalBreaks',
            legend_kwds={'loc': 'upper left'})
```

#### Top 5 Census Tracts with Highest Number of Renter-Occupied Housing Units

```python
geo_df.sort_values(by="Renter-Occupied Housing Units", ascending = False).head(5).plot.barh(x='FIPS',
                            y="Renter-Occupied Housing Units", 
                            title='Top 5 Census Tracts with Highest Number of Renter-Occupied Housing Units',
                            legend=False)
```

## Transportation Characteristics

```python
geo_df["Mode Share - Public Transit"] = geo_df["Workers 16 Years and Over: Public Transportation (Includes Taxicab)"] / geo_df["Workers 16 Years and Over:"]
```

#### Percent of Workers 16 Years and Over using Public Transit

```python
geo_df.plot(figsize=(9,9),
            column="Mode Share - Public Transit",
            legend=True, 
            scheme='NaturalBreaks',
            legend_kwds={'loc': 'upper left'})
```

#### Top 5 Census Tracts with Most Residents using Public Transit

```python
geo_df.sort_values(by="Workers 16 Years and Over: Public Transportation (Includes Taxicab)", ascending = False).head(5).plot.barh(x='FIPS',
                            y="Workers 16 Years and Over: Public Transportation (Includes Taxicab)", 
                            title='Top 5 Census Tracts with Most Residents using Public Transit',
                            legend=False)
```
